```python
#Answer 3B here; Please comment each part

def missing_row_values(inputdf):
    #made temporary dataframe "df2"
    df2 = pd.DataFrame (
        #isna() identifies nulls, sum() takes sum, with axis=1 making it rowwise
        inputdf.isna().sum(axis=1),
        #label for the column
        columns=['missing_values']
    )
    #just for me to check that I got the missing row value sums:
    #print(df2) 
    
#executing the new function, making a float to hold data
missing_values_column = missing_row_values(NHL_Goalies)

#appending the new column data to my dataframe
NHL_Goalies['missing_values'] = missing_values_column

#just for me to check that the column was appended correctly:
list(NHL_Goalies.columns.values) 

```

### Q5) GroupBy Question
##### 1. Subset the NHL_Goalies DataFrame where Injuries is not known <font color='red'>(Injuries=NaN)</font>; 
##### 2. Use a Merge with the GAA_  DF to produce a NEW DF (*called mergedDF*) 
###### *** USE A JOIN THAT PRESERVES ORDER AND USES THE INTERSECTION OF KEYS***
###### *** USE BOTH THE LEFT AND RIGHT INDEXES AS JOIN KEYS***
##### 3. From the mergedDF keep FirstName/LastName/Team/Cntry/ SV% /GA/GAA/ TOI (note one copy for any duplicate column)
##### 4. GroupBy country on mergedDF and aggregate The Means, Mins, and Maximum of the kept columns
#### 5. Create a dataframe of the <font color='red'>TOI </font>groupby object and write it to a comma seperated value file called ('TOI_2017.csv')

## Notes:

1. Review or read up pandas documentation on:
    - how to create selection criteria
    - how aggs/joins work: https://pandas.pydata.org/pandas-docs/stable/merging.html
2. https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.to_csv.html
3. https://stackoverflow.com/questions/14734533/how-to-access-pandas-groupby-dataframe-by-key
4. https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.get.html

```python
#Answer 5 here: please comments


#1. Subset of NHL Goalies with unknown injuries:
df1 = NHL_Goalies[NHL_Goalies["Injuries"].isnull()]

#2. Merge with GAA:

#dataframe with GAAs
GAA_DF = pd.DataFrame(NHL_Goalies['GAA'])
GAA_DF

#merged the two using pd.merge
mergedDF=pd.merge(df1, GAA_DF, left_index=True, right_index=True)
mergedDF

#3. From the mergedDF, keep FirstName/LastName/Team/Cntry/SV%/GA/GAA_y/TOI:

#I noticed there was no 'TOI' column in the original excel file, so I substituted TOI with "MIN" = minutes played
mergedDF1 = mergedDF.loc[:,['Last Name', 'First Name', 'Team', 'Cntry', 'SV%', 'GA', 'GAA_y', 'MIN']]
mergedDF1

#4. GroupBy country on mergedDF and aggregate the means, mins, and maximum of the kept columns
mergedDF1.groupby("Cntry").agg(['mean', 'min', 'max'])

#5. Create a dataframe of the TOI groupby object and write it to a comma separated value file called ('TOI_2017.csv')

#again, because I couldn't find the TOI column, I'll use the MIN groupby object:
TOI_minutes = pd.DataFrame(mergedDF1.groupby("Cntry").agg({'MIN': ['mean', 'min', 'max']}))
TOI_minutes

#saved the file as the name indicated, but the data used is not TOI
TOI_minutes.to_csv('TOI_2017.csv', sep='\t', encoding='utf-8')
```

# Q6) Write a function(s)  
## that subsets a dataframe by removing rows that are WITHIN the InterQuartile Region of a specified column
<br> </br>
### The function takes as input 
##### A: Pandas_Dataframe (For example use NHL_GoaliesDF)
##### B. Column name
### Add a condition that returns an error if the Column Name chosen is not:
##### A. Numeric Dtype
##### B. Does not exist in the Numeric Columns
### The output is a DataFrame that removes rows outside of the specified columns IQR
<br> </br>
<font color='red'>**You can use smaller helper functions **<font>
<br> </br>
See --> <href>=https://en.wikipedia.org/wiki/Interquartile_range</href>
#hint: There are several pandas subfunctions that can be useful

```python
#inputdataframe = name of dataframe, x = name of column; must be written as a string
import sys

def dataframeIQR (inputdataframe, x):
    
    #making a temporary dataframe
    df1 = pd.DataFrame (inputdataframe[x])
    
    #looking at what a boxplot of the current data looks like:
    #df1.boxplot()
    
    #calculate the interquartile range
    Q1 = df1.quantile(0.25)
    Q3 = df1.quantile(0.75)
    IQR = Q3 - Q1
    
    #make a new dataframe where anything that is NOT (~) above the upper quartile / below the lower quartile is excluded 
    no_outliers_df = df1[~((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR))).any(axis=1)]
    
    #print values of column that are within the IQR:
    print(no_outliers_df)
    
    #and for funsies, I can print the new boxplot to confirm removal of outliers:
    no_outliers_df.boxplot()
    
dataframeIQR(NHL_Goalies, 'GAA')

#sorry, in all honesty, I gave up trying to add the error conditions
#this assignment is a lot harder than I was expecting! 
    
```
